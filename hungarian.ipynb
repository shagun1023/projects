```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('dataset.xlsx')
df
```

```python
df.drop(74, axis = 0, inplace  = True)
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
print("Maximum Production of Animals for Slaughtering in Tonn", df[['Cattle','Pigs','Horses','Sheep','Poultry']].max())
print("-"*70)
print("Average Production of Animals for Slaughtering in Tonn", df[['Cattle','Pigs','Horses','Sheep','Poultry']].mean())
print("-"*70)
print("Minimum Production of Animals for Slaughtering in Tonn ",df[['Cattle','Pigs','Horses','Sheep','Poultry']].min())
print("-"*70)
```

```python
df.columns
```

```python
plt.figure(figsize = (20,5))

group = df.groupby('Year')['Cattle'].sum()
group.plot(kind = 'line')
plt.grid(True)
plt.title("Cattle Production for Meat Consumption Hungaria")
```

```python
plt.figure(figsize = (20,5))

group = df.groupby('Year')['Pigs'].sum()
group.plot(kind = 'line')

plt.grid(True)
plt.title("Pigs Production for Meat Consumption Hungaria")
```

```python
plt.figure(figsize = (20,5))

group = df.groupby('Year')['Poultry'].sum()
group.plot(kind = 'line')
plt.grid(True)
plt.title("Poultry Production for Meat Consumption Hungaria")
```

```python
df2 = df[df['Year'] == 1938]
print(df2)
print("-"*80)
df3 = df[df['Year'] == 1970]
print(df3)
print("-"*80)
df4 = df[df['Year'] == 2000]
print(df4)
print("-"*80)
df1 = df[df['Year'] == 2021]
print(df1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,10))

ax1 = sns.barplot(data = df1, ax=axes[0,0])
axes[0,0].set_xlabel("Year 1938")
axes[0,0].bar_label(ax1.containers[0], fmt = '%d')
axes[0,0].grid(True)
axes[0,0].set_title("Meat Productions in 1938 in Tonn")

ax3 = sns.barplot(data = df3, ax=axes[0,1])
axes[0,1].set_xlabel("Year 1970")
axes[0,1].grid(True)
axes[0,1].bar_label(ax3.containers[0], fmt = '%d')
axes[0,1].set_title("Meat Productions in 1970 in Tonn")

ax3 = sns.barplot(data = df4, ax=axes[1,0])
axes[1,0].set_xlabel("Year 2000")
axes[1,0].grid(True)
axes[1,0].bar_label(ax3.containers[0], fmt = '%d')
axes[1,0].set_title("Meat Productions in 2000 in Tonn")


ax2 = sns.barplot(data = df2, ax=axes[1,1])
axes[1,1].set_xlabel("Year 2021")
axes[1,1].grid(True)
axes[1,1].bar_label(ax2.containers[0], fmt = '%d')
axes[1,1].set_title("Meat Productions in 1921 in Tonn")

plt.show()
```

```python
plt.figure(figsize = (6,5))
df1.plot(x = 'Year', kind = 'bar')

plt.title("2021 Year's Meat Production in Hungary")
plt.legend()
plt.grid(True)
```
