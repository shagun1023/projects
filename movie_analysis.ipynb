```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_html('movie.html')
df[0]
```

```python
df = pd.DataFrame( data  = df[0])
df
```

```python
df.value_counts()
```

```python
plt.figure(figsize = (20,5))
ax = sns.barplot(x = 'Movie', y = 'India Hindi Net', data = df)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

 ***From above plot, we see that Pathan, KGF Ch.2 and Bahubali Ch.2 had the highest revenue collection in the Indian Market.***

```python
plt.figure(figsize = (20,5))
ax = sns.barplot(x = 'Movie', y = 'Worldwide', data = df)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

**South Movies espcially Tollywood Movies are Thriving the Overseas Market But Dangal alone from Bollywood has done good effort in Overseas Market.**

**What Kind of Verdicts are Popular**

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(y = 'Verdict', data = df)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.show()
```

**There are about 11 All Time Blockbusters, 16 Blockbusters and 6 Hits.**

**Co-relations between 1st Week Collections of Indian and Woldwide**

```python
correlation_data = df.iloc[:, 1:5].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Correlation between 1st Week of Indian Collections & World Wide Collections')
plt.show()
```

**1st Week of Indian Cinema Collections are more than the World-Wide Collections due to the Language Barrier of Indian Languages. More of the Indian Cinema are made in Native as well as Hindi Language Due to its Diversity in Languages, collections are effected World wide and in India too.**

**How Much of Verdicts of the Movies earned the Most Collection Worldwide**

```python
plt.style.use('ggplot')
plt.figure(figsize = (20,6))
group_data = df.groupby('Verdict')['Worldwide'].sum()
colors = ['#701A76', '#EB2410', '#9BEB10', '#1074EB', '#EB10CD', '#EB1056']
ax = group_data.plot(kind = 'bar', color = colors, alpha = 0.7)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title('Worldwide Collection v/s Verdicts')
plt.xlabel('Verdicts')
plt.ylabel('Collections')
plt.grid(True)
plt.show()
```

**All Time Blockbuster has earned most of the Collections Worldwide i.g., Bahubali 2, RRR and KGF-2 Movies. That is followed by the Blockbuster by Pathan Movie.**

**Conclusion** : The Southern Indian Movies like RRR, KGF-2 and Bahubali 1 & 2 are most popular in india and at World Market.
